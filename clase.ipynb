ARQUITECTURA CLIENTE SERVIDOR

El cliente entra a una URL intentando acceder al recurso
El navegador toma la URL y busca la IP para hacer la peticion por http
Hypertext Transfer Protocol (HTTP) -> El server recibe la peticion, la analiza y envia otra HTTP en HTML
Se renderiza la pagina -> El server siempre envia una respuesta, sea exitosa o no

LAS PRIMERAS APLICACIONES
Static apps -> Request, revisa si lo tiene el server, envia el contenido del html (Contenido dinamico)

SSR -> Server Side Rendering (.php) Se renderiza el HTML del cliente desde el servidor.
    Archivos que controlan y gestionan las peticiones, dependiento del usuario le muestra contenido personalizado generando un html generado dinamicamente, cada recurso es diferente. (Mas lento y es mas propenso a hackear)
    Se llaman apps webs tradicionales

AJAX -> Dynamics Apps
    Guarda las interacciones del cliente, evita peticiones a la db
    La pagina ase muede modificar de manera asincrona, sin necesidad de recargar toda la pagina
    Se genera mas carga en el cliente, mas superficie para hack
    Envia un .js por las cookies

¿Como evitar ir al servidor cuando se cambia la ruta/pagina?

SPA -> Single page applications
    Se carga a medida que el usuario interactua, sin necesidad de cargar toda la pagina, similar a una app nativa

FRAMEWORKS

Un framework de software, es una tecnica que usa un codigo ya testeado para desarrollar encima de ella, reducir costos y tiempo de elaboracion.

El riesgo mas importante es que este puede evolucionar en una direccion que no nos sirva.

Las relaciones entre los frameworks son asimetricos, la arquitectura no es limpia siempre, puede que el desarrollo.
Hay que elegir un buen framework.

- Frameworks verticales
    Apps con dominio, utiles si son para un solo carril mucho tiempo (Ya tienen funciones).

- Frameworks horizontales
    Son utiles para el desarroollo de apps de multiples dominios
    No tienen funcionalidades desde el inicio

    -> Opinionated: Definen arquitectura, funciones del directorio (Django).
    -> Unopinionated: Minimalistas, no toman opiniones, nos dejan solo el sistema de peticion y respuesta.

MODELO VISTA-CONTROLADOR 
Vista -> Controlador -> Modelo
- El user interactua con la vista, vista que esta mapeada, manda peticiones de URL.
- El controlador recibe las peticiones pidiendo la informacion de un producto o pagina, este va al modelo por la DB.
- El modelo hace la consulta a la DB, cuando encuentra la peticion del controlador, envia el HTML para renderizar en el controlador o en su defecto por la Vista.
