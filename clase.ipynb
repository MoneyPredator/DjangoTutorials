El Lazy Loading es una técnica en la que los recursos o datos se cargan solo cuando son necesarios, es decir, de manera diferida. En lugar de cargar todo de una vez al inicio, se pospone la carga hasta el momento en que el recurso o dato es requerido.

    - ¿Cómo funciona?
Inicialización: Cuando se carga una aplicación o página, solo se cargan los elementos esenciales.
Carga bajo demanda: Los recursos adicionales (como imágenes, datos de una API, módulos de código, etc.) se cargan solo cuando el usuario los necesita o cuando están a punto de ser mostrados.
Optimización: Esto reduce el tiempo de carga inicial y mejora el rendimiento, especialmente en aplicaciones con muchos recursos.

    - Ejemplos de uso:
Imágenes en una página web: Las imágenes fuera de la pantalla (no visibles inicialmente) se cargan solo cuando el usuario hace scroll y están a punto de aparecer.
Módulos de una aplicación: En frameworks como Angular o React, los módulos de la aplicación se cargan solo cuando el usuario navega a una ruta específica.
Datos en una base de datos: En ORMs (Object-Relational Mappers) como Entity Framework (en .NET) o Hibernate (en Java), las relaciones entre tablas no se cargan hasta que se accede a ellas.

    - Ventajas:
Mejora el rendimiento inicial.
Reduce el consumo de memoria y ancho de banda.
Ideal para aplicaciones con muchos recursos o datos.

    - Desventajas:
Puede haber un pequeño retraso cuando se necesita el recurso.
Requiere una implementación cuidadosa para evitar problemas de usabilidad.


**Eager Loading (Carga ansiosa)**
El Eager Loading es lo opuesto al Lazy Loading. En esta técnica, todos los recursos o datos se cargan de inmediato, tan pronto como se inicia la aplicación o se realiza una consulta.

    - ¿Cómo funciona?
Carga completa: Cuando se realiza una operación (por ejemplo, una consulta a una base de datos), no solo se cargan los datos principales, sino también todos los datos relacionados.
Precarga: Todos los recursos están disponibles desde el principio, sin necesidad de cargarlos posteriormente.

    Ejemplos de uso:
Consultas a bases de datos: Si tienes una tabla de "Usuarios" y otra de "Pedidos", con Eager Loading, al cargar un usuario, también se cargan todos sus pedidos relacionados de inmediato.
Recursos en una aplicación: En una aplicación web, todas las imágenes, estilos y scripts se cargan al inicio.

    - Ventajas:
Los datos o recursos están disponibles de inmediato cuando se necesitan.
Evita retrasos en la carga posterior.

    - Desventajas:
Mayor tiempo de carga inicial.
Consumo innecesario de memoria y ancho de banda si no todos los recursos se utilizan.

## **Fakers**

Los **Fakers** son bibliotecas o herramientas que permiten generar datos falsos o aleatorios, pero realistas, para simular información en una aplicación. Estos datos son útiles para pruebas, desarrollo o para llenar una base de datos con información de ejemplo.

### ¿Para qué sirven los Fakers?

- **Pruebas**: Permiten probar funcionalidades sin necesidad de datos reales.
- **Desarrollo**: Ayudan a simular escenarios reales durante el desarrollo de una aplicación.
- **Datos de ejemplo**: Son ideales para llenar una base de datos con información ficticia pero coherente.

### Características de los Fakers:
- Generan datos realistas, como nombres, direcciones, correos electrónicos, números de teléfono, etc.
- Son personalizables para adaptarse a las necesidades específicas de tu aplicación.
- Suelen estar disponibles en múltiples lenguajes de programación (JavaScript, Python, PHP, etc.).

### Ejemplos de bibliotecas Faker:
- **JavaScript**: `faker.js` (aunque ahora se recomienda usar alternativas como `@faker-js/faker`).
- **Python**: `Faker` (una biblioteca muy popular).
- **PHP**: `Faker` (integrado en frameworks como Laravel).
- **Ruby**: `Faker` (ampliamente utilizado en Ruby on Rails).

### Ejemplo de uso en Python:
```python
from faker import Faker

fake = Faker()

# Generar datos falsos
print(fake.name())       # Ejemplo: "John Doe"
print(fake.email())      # Ejemplo: "john.doe@example.com"
print(fake.address())    # Ejemplo: "123 Main St, Springfield, IL"
```

## **Patrones de Diseño**

Los **patrones de diseño** son soluciones probadas y reutilizables para problemas comunes en el diseño de software. No son código específico, sino guías o plantillas que ayudan a estructurar el código de manera eficiente y mantenible.

### Tipos de Patrones de Diseño

1. **Patrones Creacionales**:
   - Se enfocan en cómo se crean los objetos.
   - Ejemplos:
     - **Singleton**: Asegura que una clase tenga solo una instancia.
     - **Factory Method**: Define una interfaz para crear objetos, pero permite a las subclases decidir qué clase instanciar.
     - **Abstract Factory**: Proporciona una interfaz para crear familias de objetos relacionados.
     - **Builder**: Separa la construcción de un objeto complejo de su representación.
     - **Prototype**: Permite clonar objetos existentes sin hacer que el código dependa de sus clases.

2. **Patrones Estructurales**:
   - Se enfocan en cómo se componen las clases y objetos para formar estructuras más grandes.
   - Ejemplos:
     - **Adapter**: Permite que clases con interfaces incompatibles trabajen juntas.
     - **Composite**: Compone objetos en estructuras de árbol para representar jerarquías.
     - **Proxy**: Proporciona un sustituto o marcador de posición para otro objeto.
     - **Facade**: Proporciona una interfaz simplificada a un conjunto de interfaces en un subsistema.
     - **Decorator**: Añade funcionalidad a un objeto de manera dinámica.

3. **Patrones de Comportamiento**:
   - Se enfocan en la interacción y responsabilidades entre objetos.
   - Ejemplos:
     - **Observer**: Define una dependencia uno a muchos entre objetos para que cuando uno cambie, todos sean notificados.
     - **Strategy**: Permite definir una familia de algoritmos, encapsular cada uno y hacerlos intercambiables.
     - **Command**: Encapsula una solicitud como un objeto.
     - **State**: Permite que un objeto altere su comportamiento cuando su estado interno cambia.
     - **Template Method**: Define el esqueleto de un algoritmo en una operación, delegando algunos pasos a las subclases.

## **Normalización de Modelos**

La **normalización de modelos** es un proceso utilizado en el diseño de bases de datos relacionales para organizar los datos de manera eficiente y reducir la redundancia. Consiste en aplicar una serie de reglas (llamadas formas normales) para estructurar las tablas y sus relaciones.

### Objetivos de la Normalización:
- Eliminar la redundancia de datos.
- Evitar anomalías en la inserción, actualización y eliminación de datos.
- Mejorar la integridad y consistencia de los datos.

### Formas Normales (NF - Normal Forms)

1. **Primera Forma Normal (1NF)**:
   - Cada columna debe contener valores atómicos (indivisibles).
   - Cada fila debe ser única (usando una clave primaria).

2. **Segunda Forma Normal (2NF)**:
   - Debe estar en 1NF.
   - Todos los atributos no clave deben depender completamente de la clave primaria.

3. **Tercera Forma Normal (3NF)**:
   - Debe estar en 2NF.
   - Los atributos no clave no deben tener dependencias transitivas (es decir, no deben depender de otros atributos no clave).

4. **Forma Normal de Boyce-Codd (BCNF)**:
   - Una versión más fuerte de 3NF.
   - Cada determinante debe ser una clave candidata.

5. **Cuarta Forma Normal (4NF)**:
   - Debe estar en BCNF.
   - Elimina dependencias multivaluadas.

6. **Quinta Forma Normal (5NF)**:
   - Debe estar en 4NF.
   - Elimina dependencias de reunión (join dependencies).

---

## **Relación entre Patrones de Diseño y Normalización**

- **Patrones de Diseño**: Se aplican en el diseño de software para resolver problemas comunes de arquitectura y comportamiento.
- **Normalización**: Se aplica en el diseño de bases de datos para optimizar el almacenamiento y manejo de datos.

Ambos conceptos buscan mejorar la estructura, eficiencia y mantenibilidad, pero en diferentes contextos (software vs. bases de datos).

# Preguntas y Respuestas sobre Django

## ¿Dónde se almacenan los controladores, vistas, y modelos en Django?

En Django, los controladores (vistas) y los modelos se almacenan en archivos dentro de una aplicación de Django. Los modelos generalmente se definen en el archivo `models.py`, y las vistas se definen en el archivo `views.py`. Los controladores en Django son manejados principalmente por las vistas.

## ¿Para qué sirve el archivo `settings.py`?

El archivo `settings.py` en Django es utilizado para configurar toda la aplicación. Contiene configuraciones como la base de datos a utilizar, aplicaciones instaladas, middlewares, configuraciones de internacionalización, archivos estáticos, entre otros. Es un archivo central para la configuración del proyecto.

## ¿Para qué sirven las rutas en Django?

Las rutas en Django sirven para mapear URLs a vistas específicas. Se definen en el archivo `urls.py` y permiten que Django sepa qué vista debe ejecutarse cuando un usuario visita una URL específica. Las rutas pueden incluir parámetros que se pasan a las vistas.

## ¿Cómo se deben definir los controladores en Django?

En Django, los controladores se definen como funciones o clases en el archivo `views.py`. Las funciones de vista reciben un objeto `HttpRequest` y devuelven un objeto `HttpResponse`. Las vistas basadas en clases heredan de clases como `View`, `TemplateView`, `DetailView`, etc., y definen métodos como `get` o `post` para manejar solicitudes HTTP.

## ¿Qué es un Template?

Un Template en Django es un archivo que define la estructura de la presentación de una página web. Los templates permiten separar la lógica de la presentación y suelen estar escritos en HTML con adiciones de la sintaxis de Django para incluir datos dinámicos. Los templates se almacenan generalmente en la carpeta `templates` de una aplicación.

# Preguntas y Respuestas sobre Django

## ¿Qué son los modelos?

Los modelos en Django son clases de Python que representan la estructura de una tabla en la base de datos. Cada atributo de la clase representa un campo en la tabla. Los modelos permiten interactuar con la base de datos utilizando código Python en lugar de SQL directamente.

## ¿Cuántas tablas representa un modelo?

Un modelo en Django generalmente representa una tabla en la base de datos. Cada instancia de la clase del modelo corresponde a una fila en esa tabla.

## ¿Qué son las migraciones?

Las migraciones en Django son archivos que contienen instrucciones para modificar la estructura de la base de datos. Permiten actualizar el esquema de la base de datos de manera controlada y reversible, sincronizando los cambios realizados en los modelos con la base de datos.

## ¿Para qué usar los fakers?

Los fakers se utilizan para generar datos falsos o de prueba. Son útiles para llenar bases de datos con datos de prueba durante el desarrollo o para pruebas unitarias, permitiendo simular datos realistas sin necesidad de ingresarlos manualmente.

## ¿Qué es eager loading y lazy loading?

- **Eager loading**: Es una técnica donde se cargan todos los datos relacionados de antemano, reduciendo el número de consultas a la base de datos. Es útil para optimizar el rendimiento cuando se sabe que se necesitarán los datos relacionados.
- **Lazy loading**: Es una técnica donde los datos relacionados se cargan solo cuando se accede a ellos. Esto puede reducir la carga inicial, pero puede resultar en múltiples consultas a la base de datos si se accede a muchos datos relacionados.

## ¿Cómo definir relaciones entre clases en Django?

En Django, las relaciones entre modelos se definen utilizando campos especiales como `ForeignKey` para relaciones muchos-a-uno, `ManyToManyField` para relaciones muchos-a-muchos, y `OneToOneField` para relaciones uno-a-uno. Estos campos se definen en los modelos y Django maneja la creación de las relaciones en la base de datos.

## ¿Dónde realizar las validaciones de los valores para creación de objetos?

Las validaciones de los valores para la creación de objetos en Django se pueden realizar en varios lugares:
- En el modelo, utilizando el método `clean` o validaciones de campo.
- En los formularios, utilizando la lógica de validación de Django Forms.
- En las vistas, antes de guardar el objeto en la base de datos.

6. Estadísticas de Vuelos (25%)
En views.py, crea la vista para mostrar las estadísticas.

python
Copy

```python
from django.db.models import Avg
from django.shortcuts import render
from .models import Flight

def estadisticas_vuelos(request):
    nacionales = Flight.objects.filter(tipo='Nacional').count()
    internacionales = Flight.objects.filter(tipo='Internacional').count()
    promedio_nacionales = Flight.objects.filter(tipo='Nacional').aggregate(Avg('precio'))['precio__avg'] or 0

    return render(request, 'estadisticas_vuelos.html', {
        'nacionales': nacionales,
        'internacionales': internacionales,
        'promedio_nacionales': promedio_nacionales,
    })
```

 Arquitectura Orientada a Servicios
- Que es un servicio? 
  Pueden servir como intermedio entre framework y no desarrollo
    Un servicio en una funcion definida que no depende de contexto, estado u otros servicios.
    Representacion logica de una actividad de negocios repetible, se puede saber como se usa y no como funciona
    - Comunicacion con servicios   
        SOAP (Single Object Access Protocol) -> Protocolo de cominicacion en los servicios web de tecnologia, los consumidores de rervicios y los proveedores envian request/reply por mensajes XML generalmente por HTTP

        REST (Representantional State Transfer) -> Un servicio en el que el consumidor envia request HTTP, generalmente se maneja usando JSON
    Los servicios no hacen todo el trabajo
API KEY (Licencia para servicios (Application Programming Interface Key)) -> Invoa los servicios, identificador unico, se usa para asistencia en 
        (i) Autenticacion
        (ii) Tracking
        (iii) Controlar como el servicio esta siendo usado, los derechos de servicio

SOA (Service Oriented Architecture) -> Patron de arquitectura
- Permite la composicion: Se puede comunicar y ensamblar con otros servicios y crear otros mas complejos
- Deben poder ser descubiertos: Deben proveee metadatos los cuales se puedan descubrir e interpretar de manera efectiva
- Especifica contrato: Especifica que datos envia, recibe y entrega.
    Tienen 2 rokes
    - Proveedor: Es quien mantiene el servicio y quien lo hace disponible a otros para ser usado
    - Consumidor: Localiza los metadatos del servicio y construye los componentes necesarios del cliente para "atar"

Patrones de Diseño en Python

            *Singleton*: Es una clase con instancia y un punto de acceso bien definido -> Garantiza que cada que se cree una instancia de clase, se obtenga la misma instancia

            *Prototype*: Soporta crearse a si mismo

            *Builder*: Separa la construccion de un objeto de su representacion, haciendo que el proceso de construcicion pueda ser usado en la construccion de diferentes representaciones

    - Estructurales; COmbinar clases u objetos en estructuras mas grandes, las cuales son mas que la suma de sus partes
            *Adapter*: Envuelve o adapta una immplementacion

            *Facade*: Permite crear una interfaz unificada a partir de multiples interfaces de un subsistema, crea una fachada simple de un subsistema mas complejo.

    - Conductuales: Como se comportan las distintas estructuras, su composicion se hace en tiempo de ejecucion, cuando se considera un patron conductual se consideran guias

    Cuando hablamos REST FULL o REST-like, tienen similitudes, toda la estructura gira en torno a URL, y las interacciones van con un verbo HTTP GET /cats/15 versus DELETE cats/15

        -> CLiente-Servidor
        -> Stateless
        -> Cacheables
        -> Sistema por capas: Jerarquias para gestionar la complegidad y mejorar la escalabilidad
        -> Code on demand: Permitir enviar del server al cliente
        -> Interfaz Uniforme: Restricciones fundamentales para desacoplar arquitectura, todo se acomode a la interfaz.
